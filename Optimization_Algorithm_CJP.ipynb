# **Candy Jar Placement Optimization Problem**

Problem Statement: We must find the optimal coordinates for a given candy jar to place on a table of finite length. This optimal location must also be equidistant from a set of 4 given objects placed around said candy jar. Locations of objects are given as follows: 

image.png

The domain of possible coordinates for candy jar are as follows:

image.png

We are also given that the radius of the candy jar is 2.5 inches

```python
#import statements
import numpy as np
from scipy.optimize import minimize
```

```python
#objective function that inputs current coordinates of candy jar and finds total distance from all four objects
def obj_funct(x):

    x_coords = [56, 156, 60, -132] #X coordinates of all four objects
    y_coords = [-156, 33, 120, 0] #Y coordinates of all four objects
    
    dist_list = [] #list of distances from each object

    #adding distance from each point by using np.linalg.norm to dist_list
    for i in range(4):
        dist_list.append(np.linalg.norm(np.array([x_coords[i] - x[0], y_coords[i] - x[1]])))

    return sum(dist_list) #returns sum of distances (goal is to minimize this value)
```

```python
bnds = ((-37.0, 37.0), (-21.0, 21.5)) #domain of where coordinates of origin of candy jar can be

sol = minimize(obj_funct, (0.0,0.0), method='SLSQP', bounds=bnds)
sol #X value outputted is optimal candy jar location
```
